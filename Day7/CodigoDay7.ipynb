```python

# Figure width and height
fig, ax = plt.subplots(figsize=(13, 13), constrained_layout=False)

layers = plot(
    # Coordinates or location name
    (-16.5020205, -68.1332655), radius=1500,
    ax=ax,
    layers={
        'perimeter': {'fc': '#F2F4CB', 'ec': '#dadbc1', 'lw': 0, 'hatch': 'ooo...', 'zorder': 0},
        'streets': {
            'circle': True,
            'custom_filter': '["highway"~"motorway|trunk|primary|secondary|tertiary|residential|service|unclassified|pedestrian|footway"]',
            'width': {
                'motorway': 5,
                'trunk': 5,
                'primary': 4.5,
                'secondary': 4,
                'tertiary': 3.5,
                'residential': 3,
                'service': 2,
                'unclassified': 2,
                'pedestrian': 2,
                'footway': 1,
            },
            'fc': '#2F3737', 'ec': '#475657', 'alpha': 1, 'lw': 0, 'zorder': 3,
        },
        'building': {
            'circle': True,
            'tags': {'building': True, 'landuse': 'construction'},
            'union': True,
            'palette': ['#FFC857', '#E9724C', '#C5283D'],
            'ec': '#2F3737', 'lw': 0.5, 'zorder': 4,
        },
        'water': {
            'circle': True,
            'tags': {'natural': ['water', 'bay']},
            'fc': '#a1e3ff', 'ec': '#2F3737', 'hatch': 'ooo...', 'hatch_c': '#85c9e6', 'lw': 1, 'zorder': 2,
        },
        'green': {
            'circle': True,
            'tags': {'landuse': 'grass', 'natural': ['island', 'wood'], 'leisure': 'park'},
            'fc': '#D0F1BF', 'ec': '#2F3737', 'lw': 1, 'zorder': 1,
        },
        'forest': {
            'circle': True,
            'tags': {'landuse': 'forest'},
            'fc': '#64B96A', 'ec': '#2F3737', 'lw': 1, 'zorder': 1,
        },
        'parking': {
            'circle': True,
            'tags': {'amenity': 'parking', 'highway': 'pedestrian', 'man_made': 'pier'},
            'fc': '#F2F4CB', 'ec': '#2F3737', 'lw': 1, 'zorder': 3,
        },
    }
)

# Añadir título
plt.title('CITY OF LA PAZ - BOLIVIA', fontsize=17, fontweight='bold', x=0.18, y=0.93, color='#ffffff',
          bbox=dict(facecolor='black', alpha=0.8, edgecolor='black', boxstyle='round,pad=0.7'))

# Añadir créditos personalizados
plt.text(0.9, 0.06, 'Source: OpenStreetMap / Credtit: Github.com/marceloprates/prettymaps / @Jorge A. Zampieri',
         fontsize=9, color='black', fontstyle='italic', ha='right', va='bottom',
         bbox=dict(facecolor='none', alpha=0.8, edgecolor='none', boxstyle='round,pad=0.5'),
         transform=ax.transAxes)

plt.savefig('../content/sample_data/rome.png', dpi=400)
plt.show()

#plt.savefig('../prints/rome.svg')
```
