```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
# Crear figura para 3 mapas en orden vertical
fig = plt.figure(figsize=(10, 15))

# Mapa 1: Plate Carree
ax1 = fig.add_subplot(311, projection=ccrs.PlateCarree())
ax1.set_title("Proyeccin PlateCarree", fontsize=14)
ax1.add_feature(cfeature.BORDERS, linestyle=':')
ax1.add_feature(cfeature.LAND, edgecolor='black', facecolor='lightgray')
ax1.add_feature(cfeature.COASTLINE)
ax1.gridlines(draw_labels=True)

# Mapa 2: Mollweide
ax2 = fig.add_subplot(312, projection=ccrs.Mollweide())
ax2.set_title("Proyeccin Mollweide", fontsize=14)
ax2.add_feature(cfeature.BORDERS, linestyle=':')
ax2.add_feature(cfeature.LAND, edgecolor='black', facecolor='lightblue')
ax2.add_feature(cfeature.COASTLINE)
ax2.gridlines()

# Mapa 3: Interrupted Goode Homolosine
ax3 = fig.add_subplot(313, projection=ccrs.InterruptedGoodeHomolosine())
ax3.set_title("Proyeccin Interrupted Goode Homolosine", fontsize=14)
ax3.add_feature(cfeature.BORDERS, linestyle=':')
ax3.add_feature(cfeature.LAND, edgecolor='black', facecolor='palegreen')
ax3.add_feature(cfeature.COASTLINE)
ax3.gridlines()

# Ajustar y mostrar
plt.tight_layout()
plt.show()
```

```python
# Crear figura para 3 mapas en orden vertical
fig = plt.figure(figsize=(10, 16))

# Cambiar color de fondo de la figura general
fig.patch.set_facecolor('#022952')  # Color de fondo general

# Ttulo general
fig.suptitle("MAP PROJECTIONS", fontsize=27, fontweight='bold', y=0.95, color='white', style='italic')

# Mapa 1: Interrupted Goode Homolosine
ax1 = fig.add_subplot(311, projection=ccrs.InterruptedGoodeHomolosine())
ax1.set_title("Proyeccin Interrupted Goode Homolosine", fontsize=14, color='white')
ax1.set_facecolor('#04a4ac')  # Fondo del mapa
ax1.add_feature(cfeature.BORDERS, linestyle='-', linewidth=0.8, edgecolor='white')
ax1.add_feature(cfeature.LAND, edgecolor='white', facecolor='#511204')
ax1.add_feature(cfeature.COASTLINE, linewidth=1.0, edgecolor='white')
gridlines1 = ax1.gridlines(draw_labels=True, color='#b8b6b9', linestyle='--', linewidth=0.5)
gridlines1.xlabel_style = {'color': '#b8b6b9', 'fontsize': 8}  # Estilo de coordenadas X
gridlines1.ylabel_style = {'color': '#b8b6b9', 'fontsize': 8}  # Estilo de coordenadas Y

# Mapa 2: Mollweide
ax2 = fig.add_subplot(312, projection=ccrs.Mollweide())
ax2.set_title("Proyeccin Mollweide", fontsize=14, color='white')
ax2.set_facecolor('#1591c2')  # Fondo del mapa
ax2.add_feature(cfeature.BORDERS, linestyle='-', linewidth=0.5, edgecolor='white')
ax2.add_feature(cfeature.LAND, edgecolor='white', facecolor='#083004')
ax2.add_feature(cfeature.COASTLINE, linewidth=1.0, edgecolor='white')
gridlines2 = ax2.gridlines(draw_labels=True, color='#b8b6b9', linestyle='--', linewidth=0.5)
gridlines2.xlabel_style = {'color': '#b8b6b9', 'fontsize': 8}
gridlines2.ylabel_style = {'color': '#b8b6b9', 'fontsize': 8}

# Mapa 3: Interrupted Equidisctant Conic
ax3 = fig.add_subplot(313, projection=ccrs.EquidistantConic())
ax3.set_title("Proyeccin EquidistantConic", fontsize=14, color='white')
ax3.set_facecolor('#5075ff')  # Fondo del mapa
ax3.add_feature(cfeature.BORDERS, linestyle='-', linewidth=0.5, edgecolor='white')
ax3.add_feature(cfeature.LAND, edgecolor='white', facecolor='#280347')
ax3.add_feature(cfeature.COASTLINE, linewidth=1.0, edgecolor='white')
gridlines3 = ax3.gridlines(draw_labels=True, color='#b8b6b9', linestyle='--', linewidth=0.5)
gridlines3.xlabel_style = {'color': '#b8b6b9', 'fontsize': 8}
gridlines3.ylabel_style = {'color': '#b8b6b9', 'fontsize': 8}

# Comentario en la parte inferior derecha
fig.text(0.95, 0.01, "Source: Natural Earth Feature / #30DayMapChallenge / Python:@Jorge A. Zampieri", fontsize=11, color='white', style='italic', ha='right')

# Ajustar y mostrar
plt.tight_layout(rect=[0, 0.02, 1, 0.93])  # Ajusta para que no se sobrepongan los elementos
#plt.savefig('Cartography.png', dpi=400)

plt.show()
```
